# Visualisation de données tabulaires en python

Nous avons vu comment effectuer des calculs simples. Mais l'importantne sont pas tant les valeurs numérique que le sens qui y est contenu. La visualisation des données joue alors un rôle décisif.

L'écosystème de la visualisation en python est très riche, mais la librairie de référence est `matplotlib`. Cette librairie est très puissante mais assez complexe. Nous allons aborder rapidement son utilisation pour des cas simple.

```python
import numpy as np
import matplotlib.pyplot as plt

# figures affiché automatiquement dans les notebooks
%matplotlib inline
```

```python
data = np.loadtxt("../../data/temperatures_ext_np.csv", delimiter=",")
hours = range(24)
```

```python
image = plt.imshow(data)
plt.colorbar()
plt.show()
```

```python
ave_temp = np.mean(data, axis=1)
ave_plot = plt.plot(hours, ave_temp)
plt.show()
```

```python
max_plot = plt.plot(hours, np.max(data, axis=1))
plt.show()
```

```python
min_plot = plt.plot(hours, np.min(data, axis=1))
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(hours, ave_temp, color="black", marker="o")
ax.plot(hours, np.max(data, axis=1), color="red", marker="o", linestyle="dotted")
ax.plot(hours, np.min(data, axis=1), color="red", marker="o", linestyle="dotted")
ax.fill_between(hours, np.min(data, axis=1), np.max(data, axis=1), alpha=.1, color="red")
ax.set_xlim(0, 23)
fig.show()
# sauvegarder le résultat
# fig.savefig("temp_avg_min_max.png")
# fig.savefig("temp_avg_min_max.pdf") ...
```

## Exercice : affichage de la moyenne avec +- l'écart type

```python
std_temp = np.std(data, axis=1)
fig, ax = plt.subplots()
ax.plot(hours, ave_temp, color="black", marker="o")
ax.plot(hours, ave_temp-std_temp, color="red", marker="o", linestyle="dotted")
ax.plot(hours, ave_temp+std_temp, color="red", marker="o", linestyle="dotted")
ax.fill_between(hours, ave_temp-std_temp, ave_temp+std_temp, alpha=.1, color="red")
ax.set_xlim(0, 23)
fig.show()
```

## Exercice : imshow et interpolation

la figure généré plus haut est très "pixelisé" du fait de la résolution faible des données. Il est possible de fournir un argument réglant l'interpolation pour lisser l'image obtenu. Inspectez la documentation de `imshow` et testez les interpolations.
